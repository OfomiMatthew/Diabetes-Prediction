Objective: Predict if a patient is diabetic(1) or not(0)

```python
#importing libraries
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#importing data, exploring and processing
data = pd.read_csv('diabetes.csv')
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Outcome'].value_counts()
```

```python
data.groupby('Outcome').mean()
```

```python
X = data.drop(columns = 'Outcome', axis = 1)
Y = data['Outcome']
```

```python
X
```

```python
Y
```

```python
#Standardize data to be in same scale
scale = StandardScaler()
```

```python
scale.fit(X)
```

```python
standardized_data = scale.transform(X)
```

```python
DataFrame(standardized_data)
```

```python
X = standardized_data
Y = data['Outcome']
```

```python
DataFrame(X)
```

```python
 Series(Y)
```

```python
#splitting into training and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y, test_size=0.2,random_state=42)
```

```python
DataFrame(X_train)
```

```python
DataFrame(X_test)
```

```python
DataFrame(Y_train)
```

```python
DataFrame(Y_test)
```

```python
#Training model using support vector classifier
classifier = svm.SVC(kernel = 'linear')
classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
Series(y_pred)
```

```python
#Accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Training data accuracy score is {}'.format(training_data_accuracy))
```

```python
#Accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Test data accuracy score is {}'.format(test_data_accuracy))
```

```python
#Making a Predictive system for a new data set
new_data = (4,110,92,0,0,37.6,0.191,30)
new_data = np.asarray(new_data)
new_data = new_data.reshape(1,-1)

std_data = scale.transform(new_data)
prediction = classifier.predict(std_data)

if (prediction[0]==0):
    print('Patient is not diabetic')
else:
    print('Patient is diabetic')

```
